```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

def plot_stabilization(data):
    plt.figure(figsize=(10, 6))

    for iteration, values in data.items():
        # Generate x values as time in milliseconds (10 ms intervals)
        x = [i * 10 for i in range(len(values))]
        plt.plot(x, values, label=iteration, marker='o', linestyle='-')

    plt.title('Stabilization Test')
    plt.xlabel('Time (ms)')
    plt.ylabel('Pressure Differential (Pa)')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

import matplotlib.pyplot as plt
import pandas as pd

def plot_compare_aggregated(title, data_key, plot_data, offset_percentage=0.20):
    '''
    Plots four subplots in a 2x2 grid comparing data across obstacles, and marks the point where the difference between 
    "with suspension" and "without suspension" is largest with a dotted line.
    The annotation for the max difference is placed below the "without suspension" value by a given percentage and centered.
    
    Parameters:
    - data_key (str): The key representing the y-axis data.
    - plot_data (list of tuples): A list where each tuple contains:
        - title (str): The title of the subplot.
        - categories (list of tuples): A list of (label, data) pairs.
    - offset_percentage (float): The percentage by which to lower the annotation (default is 5%).
    '''
    fig, axes = plt.subplots(2, 2, figsize=(14, 12), sharex=False, sharey=True)
    fig.suptitle(title, fontsize=16)
    
    # Flatten axes array for easy iteration
    axes = axes.flatten()

    for idx, (title, categories) in enumerate(plot_data):
        ax = axes[idx]
        max_y = 0  # Initialize max_y to adjust the limit
        max_x_length = 0  # Track maximum x length for setting xticks
        max_diff_index = -1  # Store index of largest difference
        
        # Store data for "with suspension" and "without suspension"
        with_suspension = None
        without_suspension = None

        for label, data in categories:
            df = pd.DataFrame(data)
            x_values = list(range(1, len(df) + 1))  # Start x-values from 1
            y_values = df[data_key].tolist()

            # Plot the line for the current label
            ax.plot(x_values, y_values, label=label, marker='o', linestyle='-')

            if "With Suspension" in label:
                with_suspension = y_values
            elif "Without Suspension" in label:
                without_suspension = y_values
                
            

            max_y = max(max_y, max(y_values))
            max_x_length = max(max_x_length, len(df))
        
        # Find the index where the difference is largest
        if with_suspension and without_suspension:
            diff = [abs(a - b) for a, b in zip(with_suspension, without_suspension)]
            max_diff_index = diff.index(max(diff))  # Get index of the largest difference
            
            # Annotate the largest difference point with a dotted line
            ax.plot([x_values[max_diff_index], x_values[max_diff_index]], 
                    [with_suspension[max_diff_index], without_suspension[max_diff_index]], 
                    'r--', label="Max Difference", linewidth=2)  # Dotted red line

            # Calculate the vertical offset as a percentage of the "without suspension" value
            offset = offset_percentage * without_suspension[max_diff_index]

            # Optionally, annotate the line with text showing the difference value below the "without suspension" point
            ax.text(x_values[max_diff_index], 
                    without_suspension[max_diff_index] - offset,  # Adjusted y position with percentage offset
                    f'  Max Diff\n({abs(with_suspension[max_diff_index] - without_suspension[max_diff_index]):.2f})',
                    color='red', fontsize=10, verticalalignment='top', horizontalalignment='center')  # Center aligned text
        
        a = np.array(with_suspension)
        b = np.array(without_suspension)
        difference = a - b
        percentage_difference = (difference / b) * 100
        np.set_printoptions(suppress=True, precision=2)
        print(title)
        print("Absolute difference:", difference)
        print("Percentage difference (%):", percentage_difference)
        print("---------------------")
        ax.set_title(title)
        ax.set_xlabel('Position')
        ax.set_ylabel(data_key.replace("Average", ""))
        ax.grid(True, linestyle='--', alpha=0.7)

        # Adjust y-axis range and x-ticks
        ax.set_ylim(0, max_y * 1.1)
        ax.set_xticks(range(1, max_x_length + 1))
        ax.tick_params(axis='x', rotation=45)  # Rotate for visibility
        
        # Add legend inside each subplot
        ax.legend(loc='upper right', fontsize='small', framealpha=0.8)

    # Adjust layout
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()




    
def plot_average_histogram(title, data_key):
    """
    Plots a grouped histogram for each obstacle type, showing 
    the average value of the specified data key with and without suspension, 
    including error bars (standard deviation).

    Parameters:
    - title (str): Title of the plot.
    - data_key (str): The key representing the data to average.
    """
    obstacle_groups = [
        ("Stairs", stairs_with_suspension, stairs_without_suspension),
        ("Half Stairs", half_stairs_with_suspension, half_stairs_without_suspension),
        ("Curve", curve_with_suspension, curve_without_suspension),
        ("Pipe", pipe_with_suspension, pipe_without_suspension)
    ]
    
    labels = []
    with_suspension_averages = []
    without_suspension_averages = []
    with_suspension_errors = []
    without_suspension_errors = []

    # Extract the average values and standard deviations
    for label, with_data, without_data in obstacle_groups:
        labels.append(label)
        with_suspension_averages.append(np.mean(with_data[data_key]))
        without_suspension_averages.append(np.mean(without_data[data_key]))
        with_suspension_errors.append(np.std(with_data[data_key]))
        without_suspension_errors.append(np.std(without_data[data_key]))

    # Define bar width and positions
    x = np.arange(len(labels))
    width = 0.35
    a = np.array(with_suspension_averages)
    b = np.array(without_suspension_averages)
    difference = a - b
    percentage_difference = (difference / b) * 100
    print("Absolute difference:", difference)
    print("Percentage difference (%):", percentage_difference)
    

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(x - width/2, with_suspension_averages, width, yerr=with_suspension_errors, 
            label='With Suspension', color='skyblue', capsize=5)
    plt.bar(x + width/2, without_suspension_averages, width, yerr=without_suspension_errors, 
            label='Without Suspension', color='salmon', capsize=5)

    # Labels and Title
    plt.title(title)
    plt.ylabel(data_key.replace("Average", ""))
    plt.xticks(x, labels)
    plt.legend(title="Categories", loc='lower right', fontsize='small', framealpha=0.8)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
    
def plot_correlation():
    # Group the data for easier plotting
    data_with_suspension = {
        "Stairs": stairs_with_suspension,
        "Half Stairs": half_stairs_with_suspension,
        "Curve": curve_with_suspension,
        "Pipe": pipe_with_suspension,
    }

    data_without_suspension = {
        "Stairs": stairs_without_suspension,
        "Half Stairs": half_stairs_without_suspension,
        "Curve": curve_without_suspension,
        "Pipe": pipe_without_suspension,
    }

    # Create subplots: 2 columns (with and without suspension) and len(data_with_suspension) rows
    fig, axes = plt.subplots(nrows=len(data_with_suspension), ncols=2, figsize=(12, 15), sharex=True, sharey=True)

    for idx, (obstacle, data) in enumerate(data_with_suspension.items()):
        # Convert to DataFrame for easy handling
        df_with = pd.DataFrame(data)
        df_without = pd.DataFrame(data_without_suspension[obstacle])

        # Calculate correlation coefficients
        corr_with = df_with['Pressure Differential (Pa)'].corr(df_with['Normal Force (N)'])
        corr_without = df_without['Pressure Differential (Pa)'].corr(df_without['Normal Force (N)'])

        # Plot with suspension
        sns.regplot(
            x='Pressure Differential (Pa)', 
            y='Normal Force (N)', 
            data=df_with,
            ax=axes[idx, 0], 
            scatter_kws={'color': 'blue', 's': 40, 'alpha': 0.6},
            line_kws={'color': 'blue'},
            label='With Suspension'
        )
        axes[idx, 0].set_title(f'{obstacle} - With Suspension')
        axes[idx, 0].grid(True, linestyle='--', alpha=0.5)
        axes[idx, 0].text(0.95, 0.05, f'r = {corr_with:.2f}', 
                          transform=axes[idx, 0].transAxes, 
                          fontsize=10, verticalalignment='bottom', 
                          horizontalalignment='right', color='blue')

        # Plot without suspension
        sns.regplot(
            x='Pressure Differential (Pa)', 
            y='Normal Force (N)', 
            data=df_without,
            ax=axes[idx, 1], 
            scatter_kws={'color': 'orange', 's': 40, 'alpha': 0.6},
            line_kws={'color': 'orange'},
            label='Without Suspension'
        )
        axes[idx, 1].set_title(f'{obstacle} - Without Suspension')
        axes[idx, 1].grid(True, linestyle='--', alpha=0.5)
        axes[idx, 1].text(0.95, 0.05, f'r = {corr_without:.2f}', 
                          transform=axes[idx, 1].transAxes, 
                          fontsize=10, verticalalignment='bottom', 
                          horizontalalignment='right', color='orange')

    # Setting axis labels
    for ax in axes.flat:
        ax.set_xlabel('Pressure Differential (Pa)')
        ax.set_ylabel('Normal Force (N)')

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# Visualization and p-value display
def ttest(position):
    # Data extraction
    metrics = ["Pressure Reading (Pa)", "Normal force (N)"]

    # Create a DataFrame for visualization
    plot_data = []

    for metric in metrics:
        for group in ["With Suspension", "Without Suspension"]:
            for value in data[group][position][metric]:
                plot_data.append({
                    "Metric": metric,
                    "Group": group,
                    "Value": value
                })

    df = pd.DataFrame(plot_data)

    # Plot each metric separately
    fig, axes = plt.subplots(1, 2, figsize=(15, 4))
    fig.suptitle(f"{position}", fontsize=16)

    for i, metric in enumerate(metrics):
        ax = axes[i]

        # Extract the values for t-test
        with_suspension = data["With Suspension"][position][metric]
        without_suspension = data["Without Suspension"][position][metric]
        
        # Calculate t-test (two-tailed, unequal variance: Welch's t-test)
        t_stat, p_value = ttest_ind(with_suspension, without_suspension, equal_var=False)

        # Plot
        sns.boxplot(x='Group', y='Value', data=df[df["Metric"] == metric], palette='Set2', ax=ax)
        ax.set_ylabel(metric)
        ax.set_xlabel("")
        ax.set_title(metric)

        # Add legend manually with t and p values
        ax.legend([f"t = {t_stat:.4f}, p = {p_value:.4f}"], loc='lower left')

    plt.tight_layout(rect=[0, 0, 1, 1])  # Adjust layout to make space for the suptitle
    plt.show()
```
